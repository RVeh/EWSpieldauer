## Beispielrechnung von E(V) mit fester Spielsituation

ZunÃ¤chst berechnen wir den Erwartungswert fÃ¼r eine einfache Spielsituation, um damit das zugrundeliegende Prinzip zu entdecken.

### ðŸ§® Beispiel: Berechnung von $E(1, 1, 0)$

Gegeben seien die Wahrscheinlichkeiten:

- $p_1 = \frac{1}{2}$, $p_2 = \frac{1}{3}$, $p_3 = \frac{1}{6}$

und die Spielsituation $V = (1, 1, 0)$, d.h. es liegt je ein Chip auf Feld 1 und 2, kein Chip auf Feld 3.

ZunÃ¤chst: Was kann bei einem Wurf passieren?

- Mit Wahrscheinlichkeit $p_1$ trifft man Feld 1 â‡’ es bleibt $(0,1,0)$
- Mit Wahrscheinlichkeit $p_2$ trifft man Feld 2 â‡’ es bleibt $(1,0,0)$
- Mit Wahrscheinlichkeit $p_3$ trifft man Feld 3 â‡’ nichts passiert (da leer)

Daher ergibt sich die folgende Gleichung:

$$
E(1,1,0) = 1+p_1 \cdot E(0,1,0) + p_2 \cdot E(1,0,0) + p_3 \cdot E(1,1,0).
$$

Wir sehen: Der gesuchte Erwartungswert kommt auf beiden Seiten der Gleichung vor!  
Das liegt daran, dass **mit Wahrscheinlichkeit $p_3$ Ã¼berhaupt nichts passiert** â€“ wir mÃ¼ssen also **noch einmal wÃ¼rfeln**.

Diese Gleichung lÃ¤sst sich nach $E(1,1,0)$ auflÃ¶sen:

$$
E(1,1,0) \cdot (1 - p_3) = 1+p_1 \cdot E(0,1,0) + p_2 \cdot E(1,0,0)
$$

Dividiert man durch $1 - p_3 = p_1 + p_2$, erhÃ¤lt man:

$$
E(1,1,0) = \frac{1+p_1 \cdot E(0,1,0) + p_2 \cdot E(1,0,0)}{p_1 + p_2}
$$

$E(0,1,0)$ und $E(1,0,0)$ mÃ¼ssen nun nach dem gleichen Prinzip berechnet werden.

$$
E(0,1,0) = 1+p_1 \cdot E(0,1,0) + p_2 \cdot E(0,0,0) + p_3 \cdot E(0,1,0)
$$

Damit folgt:

$$
E(0,1,0)(1-p_1-p_3) = 1+p_2 \cdot E(0,0,0).
$$

Mit $E(0,0,0)=0$ und $p_2=\frac{1}{3}$ ergibt sich 

$$
E(0,1,0)=\frac{1}{p_2}=3
$$

Eine entsprechende Berechnung fÃ¼r $E(1,0,0)$ liefert 

$$
E(1,0,0)=\frac{1}{p_1}=2
$$

Zusammengefasst ergibt sich
$$ 
E(1,1,0)=\frac{1+\frac{1}{2}\cdot 3+\frac{1}{3}\cdot 2}{\frac{5}{6}}=\frac{19}{5}
$$

### ðŸŽ¯ Verallgemeinerung: Rekursive Definition von $ E(V) $

Wir betrachten eine Spielsituation $ V = (v_1, \dots, v_m) $, in der ein Spieler Chips auf $m$  Felder verteilt hat. In jedem Wurf wird mit Wahrscheinlichkeit $p_j$ das Feld $j$ getroffen. Liegt dort mindestens ein Chip, wird er entfernt.

Die erwartete Anzahl an WÃ¼rfen $E(V)$, bis alle Chips abgerÃ¤umt sind, erfÃ¼llt die folgende rekursive Beziehung:

$$
E(V) =
\begin{cases}
0, & \text{falls } V = (0, \dots, 0) \\
\displaystyle{\frac{1}{s} + \frac{1}{s}\sum_{j=1}^m q_j \cdot E(V - e_j)}, & \text{sonst}
\end{cases}
$$

Dabei gilt:
- $ \mathbf{1}_{\{v_j > 0\}} $ ist die Indikatorfunktion, die prÃ¼ft, ob auf Feld $j$ noch ein Chip liegt.
- $ q_j := p_j \cdot \mathbf{1}_{\{v_j > 0\}}$ liefert die Wahrscheinlichkeit $p_j$, falls auf dem $j$-ten Feld mindestens ein Chip liegt,
- $s := \sum_{j=1}^m q_j$ ist die Gesamtwahrscheinlichkeit, dass ein Wurf Ã¼berhaupt etwas bewirkt.
- $ V - e_j $ bezeichnet die Spielsituation nach einem Treffer auf Feld $j$.


---

**Bemerkung:**  
Diese Betrachtung ist ein Beispiel fÃ¼r einen **bedingten Erwartungswert**:  

Wir berechnen den Mittelwert **unter der Bedingung**, dass der Wurf Ã¼berhaupt etwas verÃ¤ndert.
Dies zeigt bereits, wie die Formel rekursiv auf einfachere Situationen zurÃ¼ckgreift.

## ðŸ” Experimentiere selbst

Hier kannst du eigene Wahrscheinlichkeiten und Verteilungen ausprobieren.

```python
from fractions import Fraction
from functools import lru_cache

# Eingabe: Anzahl der Felder
m = int(input("Anzahl der Felder (z.â€¯B. 3): "))

# Eingabe: Wahrscheinlichkeiten als BrÃ¼che
print(f"Gib {m} Wahrscheinlichkeiten als BrÃ¼che ein (z.â€¯B. 1/2):")
p = [Fraction(input(f"p_{i+1} = ")) for i in range(m)]

# Normierung der Wahrscheinlichkeiten (optional)
if sum(p) != 1:
    print("Hinweis: Wahrscheinlichkeiten wurden normalisiert.")
    p = [x / sum(p) for x in p]

# Eingabe: Verteilung der Chips
print(f"Gib {m} Chipzahlen (getrennt durch Leerzeichen) ein:")
V = tuple(map(int, input("V = ").split()))

# Rekursive Funktion 
@lru_cache(maxsize=None)
def expected_value(V, p):
    if all(v == 0 for v in V):
        return Fraction(0)
    total = Fraction(0)
    s = Fraction(0)
    for j, vj in enumerate(V):
        if vj > 0:
            Vj = list(V)
            Vj[j] -= 1
            total += p[j] * expected_value(tuple(Vj), p)
            s += p[j]
    return 1/s + total/s

# Berechnung
ev_exact = expected_value(V, tuple(p))
ev_float = float(ev_exact)

# Ausgabe
print(f"\nErwartungswert E{V} = {ev_exact} â‰ˆ {ev_float:.5f}")
```

## ðŸ”¢ Alle Erwartungswerte $E(V)$ fÃ¼r gegebene Chipanzahl $n$

FÃ¼r eine feste Anzahl an Feldern $m$ und Chips $n$ erzeugen wir alle mÃ¶glichen Verteilungen $V = (v_1, \dots, v_m)$ mit $\sum v_j = n$.

FÃ¼r jede Verteilung berechnen wir den Erwartungswert $E(V)$ exakt (als Bruch) und nÃ¤herungsweise (als Dezimalzahl), speichern sie sortiert in einer CSV-Datei, und stellen sie grafisch dar.

ZusÃ¤tzlich zeigen wir die 5 kleinsten und 5 grÃ¶ÃŸten Erwartungswerte in tabellarischer Form.

Die verwendete Formel basiert auf der rekursiven Definition:

$$
E(V) =
\begin{cases}
0, & \text{falls } V = (0, \dots, 0) \\
\displaystyle{\frac{1}{s} + \frac{1}{s}\sum_{j=1}^m q_j \cdot E(V - e_j)}, & \text{sonst}
\end{cases}
$$

Dabei gilt:
- $ \mathbf{1}_{\{v_j > 0\}} $ ist die Indikatorfunktion, die prÃ¼ft, ob auf Feld $j$ noch ein Chip liegt.
- $ q_j := p_j \cdot \mathbf{1}_{\{v_j > 0\}}$ liefert die Wahrscheinlichkeit $p_j$, falls auf dem $j$-ten Feld mindestens ein Chip liegt,
- $s := \sum_{j=1}^m q_j$ ist die Gesamtwahrscheinlichkeit, dass ein Wurf Ã¼berhaupt etwas bewirkt.
- $ V - e_j $ bezeichnet die Spielsituation nach einem Treffer auf Feld $j$.


Dies zeigt anschaulich, wie stark die Setzstrategie den Erwartungswert beeinflusst.

---

```python
from fractions import Fraction
from functools import lru_cache
import itertools
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd

# Eingaben
m = int(input("Anzahl der FÃ¤cher (z.â€¯B. 3): "))
print(f"Gib {m} Wahrscheinlichkeiten als BrÃ¼che (z.â€¯B. 1/2):")
p = [Fraction(input(f"p{i+1} = ")) for i in range(m)]

# Normalisieren
if sum(p) != 1:
    p = [x / sum(p) for x in p]

n = int(input("Anzahl der Chips (z.â€¯B. 6): "))
tick_abstand = int(input("X-Tick-Abstand (z.â€¯B. 5): "))

# Erwartungswertfunktion
@lru_cache(maxsize=None)
def expected_throws(*V):
    if all(v == 0 for v in V):
        return Fraction(0)
    total = Fraction(0)
    s = Fraction(0)
    for j, vj in enumerate(V):
        if vj > 0:
            new_state = list(V)
            new_state[j] -= 1
            total += p[j] * expected_throws(*tuple(new_state))
            s += p[j]
    return (Fraction(1) + total) / s

# Verteilungen erzeugen
verteilungen = [v for v in itertools.product(range(n+1), repeat=m) if sum(v) == n]
werte = sorted([(v, expected_throws(*v)) for v in verteilungen], key=lambda x: x[1])

# DataFrame
df = pd.DataFrame([
    {"Nr.": i+1, "Verteilung": str(v), "E(V)": str(ev), "E(V) (float)": float(ev)}
    for i, (v, ev) in enumerate(werte)
])

# CSV speichern
csv_path = f"erwartungswerte_m{m}_n{n}.csv"
df.to_csv(csv_path, index=False)
print(f"\nCSV gespeichert als: {csv_path}")

# Top 5 & Bottom 5 anzeigen
print("\nðŸ”» Kleinste Erwartungswerte:")
display(df.head(5))
print("\nðŸ”º GrÃ¶ÃŸte Erwartungswerte:")
display(df.tail(5))

# Plot
x_vals = list(range(len(werte)))
labels = [str(v) for v, _ in werte]
evs = [float(ev) for _, ev in werte]

fig = plt.figure(figsize=(6, 3))
ax = fig.add_subplot(111)

anzahl_punkte = len(evs)
if anzahl_punkte < 500:
    lw = 1.0
    ms = 3
elif anzahl_punkte < 2000:
    lw = 0.5
    ms = 2
else:
    lw = 0
    ms = 1

ax.plot(x_vals, evs, color="black", marker=".", markersize=ms, linewidth=lw)
ax.set_ylabel("E(V)", fontsize=12)
ax.set_xlabel("Verteilungen", fontsize=12)
ax.set_title(f"E(V) fÃ¼r m = {m}, n = {n}", fontsize=12)
ax.grid(True, linestyle=":", linewidth=0.5)

if len(labels) <= 28:
    ax.set_xticks(x_vals)
    ax.set_xticklabels(labels, rotation=90, fontsize=7)
else:
    ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_abstand))

plt.tight_layout()
plt.show()
pdf_path = f"erwartungswerte_m{m}_n{n}.pdf"
plt.savefig(pdf_path, format="pdf", dpi=600, bbox_inches="tight")
plt.close()

print(f"PDF gespeichert als {pdf_path}")
```

## âš”ï¸ Zwei Spieler mit Strategien $V$ und $W$: Erwartete Spieldauer $E(V, W)$

In dieser Variante spielen zwei Personen gegeneinander. Jede Person verteilt $n$ Chips auf $m$ Felder:

- Spieler A wÃ¤hlt eine Verteilung $V = (v_1, \dots, v_m)$
- Spieler B wÃ¤hlt eine Verteilung $W = (w_1, \dots, w_m)$

Ein Wurf trifft Feld $j$ mit Wahrscheinlichkeit $p_j$.
Falls auf diesem Feld bei einem der Spieler ein Chip liegt, wird **ein Chip entfernt** â€“ ggf. bei beiden Spielern gleichzeitig.

---

### ðŸ›‘ Abbruchbedingungen:

Das Spiel endet, sobald eine dieser Situationen eintritt:

- $V = W$ (z.â€¯B. durch Symmetrie oder Restgleichheit)
- $\sum_j v_j = 0$ (Spieler A hat keine Chips mehr)
- $\sum_j w_j = 0$ (Spieler B hat keine Chips mehr)

Dann gilt:

$$
E(V, W) = 0
$$

---

### ðŸ” Rekursive Formel:

FÃ¼r alle anderen FÃ¤lle wird rekursiv gerechnet:

$$
E(V, W) = \frac{1}{s} + \frac{1}{s}\sum_{j=1}^m q_j \cdot E(V - \delta_j^V, W - \delta_j^W)
$$

mit:

- $q_j := p_j \cdot \mathbf{1}_{\{v_j > 0 \lor w_j > 0\}}$
- $s := \sum_{j=1}^m q_j$ ist die Gesamtwahrscheinlichkeit, dass ein Wurf Ã¼berhaupt etwas bewirkt.

- $\delta_j^V = e_j$ falls $v_j > 0$, sonst $0$
- $\delta_j^W = e_j$ falls $w_j > 0$, sonst $0$

---

### ðŸ§  Bedeutung:

- Der Ausdruck zÃ¤hlt, wie viele WÃ¼rfe **im Durchschnitt nÃ¶tig** sind, bis das Spiel (unter den genannten Bedingungen) endet.
- Die Formel berÃ¼cksichtigt nur Felder, auf denen bei mindestens einem Spieler Chips liegen â€“ sonst wirkt ein Wurf nicht.

Dies bildet den realen Spielverlauf exakt ab und liefert einen **bedingten Erwartungswert** Ã¼ber alle mÃ¶glichen Wirkungen des nÃ¤chsten Wurfs.


---
## Programm fÃ¼r E(V,W) ##
---

```python
from fractions import Fraction
from functools import lru_cache

@lru_cache(maxsize=None)
def expected_duel(V, W, p):
    # Abbruchbedingungen: einer leer oder V = W
    if V == W or all(v == 0 for v in V) or all(w == 0 for w in W):
        return Fraction(0)
    
    total = Fraction(0)
    s = Fraction(0)
    
    for j in range(len(p)):
        vj, wj = V[j], W[j]
        if vj > 0 or wj > 0:
            qj = p[j]
            
            # Neues V
            new_V = list(V)
            if vj > 0:
                new_V[j] -= 1
            
            # Neues W
            new_W = list(W)
            if wj > 0:
                new_W[j] -= 1
            
            total += qj * expected_duel(tuple(new_V), tuple(new_W), p)
            s += qj

    return (1 + total) / s
```

## Eingabe ##
---

```python
from fractions import Fraction

# Eingaben
m = int(input("Anzahl der FÃ¤cher (z.â€¯B. 3): "))

print(f"Gib {m} Wahrscheinlichkeiten als BrÃ¼che ein (z.â€¯B. 1/2):")
p = [Fraction(input(f"p_{i+1} = ")) for i in range(m)]

# Normalisierung
if sum(p) != 1:
    p = [x / sum(p) for x in p]

print(f"Gib {m} Zahlen fÃ¼r Verteilung V (Spieler A):")
V = tuple(map(int, input("V = ").split()))

print(f"Gib {m} Zahlen fÃ¼r Verteilung W (Spieler B):")
W = tuple(map(int, input("W = ").split()))

# Berechnung
evw = expected_duel(V, W, tuple(p))
print(f"\nErwartete Spieldauer E({V}, {W}) = {evw} â‰ˆ {float(evw):.5f}")
```

## ðŸŽ¯ Vergleich eines festen $V$ gegen alle mÃ¶glichen Gegenspieler $W$

In dieser Analyse fixieren wir eine Verteilung $V$ fÃ¼r Spieler A.

Dann werden alle Verteilungen $W$ betrachtet, die dieselbe Gesamtanzahl an Chips haben, also $\sum_j w_j = n$.

FÃ¼r jede mÃ¶gliche Gegnerstrategie $W$ berechnen wir den Erwartungswert $E(V, W)$ und sortieren die Ergebnisse.  
So erkennt man, gegen welche Strategien $V$ besonders schnell oder besonders langsam endet.
---

```python
from fractions import Fraction
from functools import lru_cache
import itertools
import pandas as pd

# ðŸ”§ Anzeige-Parameter
anzeige_limit = 50       # Ab wann die Anzeige gekÃ¼rzt wird
anzahl_anzeigen = 10     # Wie viele Zeilen oben/unten gezeigt werden

# Eingaben
m = int(input("Anzahl der FÃ¤cher (z.â€¯B. 3): "))
print(f"Gib {m} Wahrscheinlichkeiten als BrÃ¼che ein (z.â€¯B. 1/2):")
p = [Fraction(input(f"p{i+1} = ")) for i in range(m)]

if sum(p) != 1:
    p = [x / sum(p) for x in p]

print(f"Gib {m} Zahlen fÃ¼r die feste Verteilung V (Spieler A):")
V = tuple(map(int, input("V = ").split()))
n = sum(V)

# Rekursive Funktion fÃ¼r E(V, W)
@lru_cache(maxsize=None)
def expected_duel(V, W, p):
    if V == W or all(v == 0 for v in V) or all(w == 0 for w in W):
        return Fraction(0)
    total = Fraction(0)
    s = Fraction(0)
    for j in range(len(p)):
        vj, wj = V[j], W[j]
        if vj > 0 or wj > 0:
            qj = p[j]
            new_V = list(V)
            new_W = list(W)
            if vj > 0:
                new_V[j] -= 1
            if wj > 0:
                new_W[j] -= 1
            total += qj * expected_duel(tuple(new_V), tuple(new_W), p)
            s += qj
    return (1 + total) / s

# Alle W mit gleicher Chipanzahl erzeugen
verteilungen_W = [w for w in itertools.product(range(n+1), repeat=m) if sum(w) == n]

# Erwartungswerte berechnen und sortieren
werte = [(w, expected_duel(V, w, tuple(p))) for w in verteilungen_W]
werte.sort(key=lambda x: x[1])

# DataFrame
df_duel = pd.DataFrame([
    {"Nr.": i+1, "Gegner W": str(w), "E(V,W)": str(ev), "E(V,W) (float)": float(ev)}
    for i, (w, ev) in enumerate(werte)
])

# Ausgabe gekÃ¼rzt oder vollstÃ¤ndig
if len(df_duel) > anzeige_limit:
    print(f"\nDie Tabelle hat {len(df_duel)} EintrÃ¤ge â€“ gekÃ¼rzte Anzeige:")
    print(f"\nðŸ”» Kleinste Erwartungswerte E(V, W):")
    display(df_duel.head(anzahl_anzeigen))
    print(f"\nðŸ”º GrÃ¶ÃŸte Erwartungswerte E(V, W):")
    display(df_duel.tail(anzahl_anzeigen))
else:
    display(df_duel)

# CSV speichern
filename = f"duell_verteilung_V{V}_gegen_alle_m{m}_n{n}.csv"
df_duel.to_csv(filename, index=False)
print(f"\nCSV gespeichert als: {filename}")
```

```python
# Bewertung: Vergleich mit E(W, V)
bewertet = []
for w, ev_vw in werte:
    ev_wv = expected_duel(w, V, tuple(p))
    if ev_vw < ev_wv:
        beurteilung = "Sieg"
    elif ev_vw > ev_wv:
        beurteilung = "Niederlage"
    else:
        beurteilung = "Unentschieden"
    bewertet.append((w, ev_vw, ev_wv, beurteilung))

# Neuer DataFrame mit Bewertung
df_bewertung = pd.DataFrame([
    {
        "Nr.": i+1,
        "Gegner W": str(w),
        "E(V,W)": str(ev_vw),
        "E(V,W) (float)": float(ev_vw),
        "E(W,V) (float)": float(ev_wv),
        "Bewertung": beurteilung
    }
    for i, (w, ev_vw, ev_wv, beurteilung) in enumerate(bewertet)
])

df_bewertung.to_csv(f"duell_bewertet_V{V}_m{m}_n{n}.csv", index=False)
print(f"CSV mit Bewertungen gespeichert.")
```
